# WISH results

```python
for meta in ['cow', 'water', 'gut']:
    label_df = pd.read_csv(f'data/{meta}_dataset_hic_gtdb.tsv', sep='\t')
    label = {acc: host for acc, host in zip(label_df.Phage, label_df.MAGs)}
    mag2species = {mag: species for mag, species in zip(label_df.MAGs, label_df.Species)}
    mag2genus = {mag: genus for mag, genus in zip(label_df.MAGs, label_df.Genus)}
    mag2family = {mag: family for mag, family in zip(label_df.MAGs, label_df.Family)}
    wish_df = pd.read_csv(f'Results/wish/{meta}.tsv', sep='\t')
    wish_df = wish_df.fillna('-')
    wish_df = wish_df[wish_df['Best hit among provided hosts'] != '-']
    wish_prediction = {acc.split('.fa', 1)[0]: host.split('.fa', 1)[0] for acc, host in zip(wish_df['Phage'], wish_df['Best hit among provided hosts'])}
    total = 0
    exact = 0
    species = 0
    genus = 0
    family = 0
    for acc in label:
        try:
            pred = wish_prediction[acc]
            if pred == '':
                continue
            elif pred.replace('_', ' ') == label[acc]:
                exact += 1
                total += 1
                species += 1
                genus += 1
                family += 1
            elif mag2species[pred] == mag2species[label[acc]] and mag2species[pred] != 'unknown':
                species += 1
                total += 1
                genus += 1
                family += 1
            elif mag2genus[pred] == mag2genus[label[acc]] and mag2genus[pred] != 'unknown':
                genus += 1
                total += 1
                family += 1
            elif mag2family[pred] == mag2family[label[acc]] and mag2family[pred] != 'unknown':
                family += 1
                total += 1
            else:
                total += 1
        except:
            pass
    print(f'{meta} exact match accuracy', exact/total)
    print(f'{meta} species accuracy', species/total)
    print(f'{meta} genus accuracy', genus/total)
    print(f'{meta} family accuracy', family/total)
    print(f'{meta} prediction rate', total/len(label))
```

## iPHoP results

```python
for meta in ['cow', 'water', 'gut']:
    label = pd.read_csv(f'data/{meta}_dataset_hic_gtdb.tsv', sep='\t')
    pred = pd.read_csv(f'Results/iphop/{meta}.csv')

    phage2mag = {row['Phage']: row['MAGs'] for _, row in label.iterrows()}
    label2species = {row['MAGs']: row['Species'] for _, row in label.iterrows()}
    label2genus = {row['MAGs']: row['Genus'] for _, row in label.iterrows()}
    label2family = {row['MAGs']: row['Family'] for _, row in label.iterrows()}


    phage2pred = {}
    for group in pred.groupby('Virus'):
        virus = group[0]
        df = group[1]
        flag = 0
        for host in df['Host genome']:
            if meta in host:
                if host.split(f'{meta}_')[1] in label['MAGs'].values:
                    phage2pred[virus] = host.split(f'{meta}_')[1]
                    flag = 1
                    break
        if flag == 0:
            try:
                phage2pred[virus] = df['Host genome'].values[0].split('_', 1)[1]
            except IndexError:
                phage2pred[virus] = df['Host genome'].values[0]

    df = pd.DataFrame({"phage": list(phage2pred.keys()), "MAGs": list(phage2pred.values())})
    ncbi2gtdb = pd.read_csv('Results/iphop/NCBITaxonomyGTDB.csv')
    ncbi2gtdb = ncbi2gtdb[ncbi2gtdb['Accession'].isin(set(df['MAGs'].unique()))]
    acc2species = {row['Accession']: row['GTDBSpecies'] for _, row in ncbi2gtdb.iterrows()}
    acc2genus = {row['Accession']: row['GTDBGenus'] for _, row in ncbi2gtdb.iterrows()}
    acc2family = {row['Accession']: row['GTDBFamily'] for _, row in ncbi2gtdb.iterrows()}
    species = []
    genus = []
    family = []

    for phage, mag in zip(df['phage'], df['MAGs']):
        if mag in acc2species:
            species.append(acc2species[mag])
        elif mag in label2species:
            species.append(label2species[mag])
        else:
            species.append('unknown')
        if mag in acc2genus:
            genus.append(acc2genus[mag])
        elif mag in label2genus:
            genus.append(label2genus[mag])
        else:
            genus.append('unknown')
        if mag in acc2family:
            family.append(acc2family[mag])
        elif mag in label2family:
            family.append(label2family[mag])
        else:
            family.append('unknown')

    df['Species'] = species
    df['Genus'] = genus
    df['Family'] = family

    family = 0
    family_total = 0
    genus = 0
    genus_total = 0
    species = 0
    species_total = 0
    cnt = 0
    total = 0
    for _, row in df.iterrows():
        mag = row['MAGs']
        phage = row['phage']
        if phage not in phage2mag:
            continue
        if label2genus[phage2mag[phage]] != 'unknown':
            if label2genus[phage2mag[phage]] == row['Genus']:
                genus += 1
            genus_total += 1
        if label2family[phage2mag[phage]] != 'unknown':
            if label2family[phage2mag[phage]] == row['Family']:
                family += 1
            family_total += 1
        if label2species[phage2mag[phage]] != 'unknown':
            if label2species[phage2mag[phage]] == row['Species']:
                species += 1
            species_total += 1
        if mag == phage2mag[phage]:
            cnt += 1
        total += 1

    print(f'{meta} exact match accuracy', cnt/total)
    print(f'{meta} species accuracy', species/species_total)
    print(f'{meta} genus accuracy', genus/genus_total)
    print(f'{meta} family accuracy', family/family_total)
    print(f'{meta} prediction rate', total/len(label))

```

```python
# water
for meta in ['gut', 'water', 'cow']:
    label_df = pd.read_csv(f'data/{meta}_dataset_hic_gtdb.tsv', sep='\t')
    label = {acc: host for acc, host in zip(label_df.Phage, label_df.MAGs)}
    cherry_df = pd.read_csv(f'Results/cherry/{meta}.tsv', sep='\t')
    cherry_df = cherry_df[cherry_df['Host'] != '-']
    cherry_prediction = {acc.split('.fa', 1)[0]: host for acc, host in zip(cherry_df.Accession, cherry_df.Host) if acc.split('.fa', 1)[0] in label}
    php_df = pd.read_csv(f'Results/php/{meta}.tsv', sep='\t')
    php_prediction = {acc.split('.fa', 1)[0]: host.split('.fa', 1)[0] for acc, host in zip(php_df.queryVirus, php_df._maxScoreHost) if acc.split('.fa', 1)[0] in label}
    phist_df = pd.read_csv(f'Results/phist/{meta}.csv')
    phist_df.dropna(inplace=True)
    phist_prediction = {acc.split('.fa', 1)[0]: host.split('.fa', 1)[0] for acc, host in zip(phist_df.phage, phist_df.host) if acc.split('.fa', 1)[0] in label}


    union = list(set(cherry_prediction) | set(php_prediction) | set(phist_prediction))
    # generate a majority vote for union prediction
    union_prediction = {}
    for item in union:
        if item in cherry_prediction and item in php_prediction and item in phist_prediction:
            votes = [cherry_prediction[item], php_prediction[item], phist_prediction[item]]
            union_prediction[item] = max(set(votes), key=votes.count)
        elif item in cherry_prediction and item in php_prediction:
            votes = [cherry_prediction[item], php_prediction[item]]
            union_prediction[item] = max(set(votes), key=votes.count)
        elif item in cherry_prediction and item in phist_prediction:
            votes = [cherry_prediction[item], phist_prediction[item]]
            union_prediction[item] = max(set(votes), key=votes.count)
        elif item in phist_prediction and item in php_prediction:
            votes = [phist_prediction[item], php_prediction[item]]
            union_prediction[item] = max(set(votes), key=votes.count)
        elif item in php_prediction:
            votes = [php_prediction[item]]
            union_prediction[item] = max(set(votes), key=votes.count)
        elif item in cherry_prediction:
            votes = [cherry_prediction[item]]
            union_prediction[item] = max(set(votes), key=votes.count)
        elif item in phist_prediction:
            votes = [phist_prediction[item]]
            union_prediction[item] = max(set(votes), key=votes.count)

    total = 0
    correct = 0
    for acc in label:
        try:
            pred = union_prediction[acc]
            if pred == '':
                continue
            elif pred.replace('_', ' ') == label[acc]:
                correct += 1
                total += 1
            else:
                total += 1
        except:
            pass

    print(f"{meta} Union + majority vote prediction accuracy", correct/total)
    print(f"{meta} Union + majority vote prediction rate", total/len(label))



    joint_prediction = {}
    for item in union:
        if item in cherry_prediction and item in php_prediction and item in phist_prediction:
            votes = [cherry_prediction[item], php_prediction[item], phist_prediction[item]]
            joint_prediction[item] = max(set(votes), key=votes.count)

    total = 0
    correct = 0
    for acc in label:
        try:
            pred = joint_prediction[acc]
            if pred == '':
                continue
            elif pred.replace('_', ' ') == label[acc]:
                correct += 1
                total += 1
            else:
                total += 1
        except:
            pass

    print(f"{meta} Joint + majority vote prediction accuracy", correct/total)
    print(f"{meta} Joint + majority vote prediction rate", total/len(label))


    joint_prediction = {}
    for item in union:
        if item in cherry_prediction and item in php_prediction and item in phist_prediction:
            if cherry_prediction[item] == php_prediction[item] and  cherry_prediction[item] == phist_prediction[item]:
                joint_prediction[item] = cherry_prediction[item]


    total = 0
    correct = 0
    for acc in label:
        try:
            pred = joint_prediction[acc]
            if pred == '':
                continue
            elif pred.replace('_', ' ') == label[acc]:
                correct += 1
                total += 1
            else:
                total += 1
        except:
            pass

    print(f"{meta} Joint + consensus prediction accuracy", correct/total)
    print(f"{meta} Joint + consensus prediction rate", total/len(label))
```
