```python
from ete3 import NCBITaxa

# Initialize NCBITaxa object (only needs to be initialized once)
ncbi = NCBITaxa()

def get_genus_and_family(species_name):
    """
    Get the genus and family names for a given species using NCBI taxonomy.
    
    Args:
        species_name (str): The name of the species (e.g., "Salmonella enterica").
    
    Returns:
        tuple: A tuple containing the genus and family names (genus, family).
               If the species is not found, it returns (None, None).
    """
    try:
        # Get the taxid for the species
        taxid = ncbi.get_name_translator([species_name])
        
        if not taxid:
            #print(f"Species '{species_name}' not found in NCBI taxonomy database.")
            return None, None
        
        species_taxid = list(taxid.values())[0][0]  # Get the first TaxID
        
        # Get the lineage for the taxid
        lineage = ncbi.get_lineage(species_taxid)
        
        # Get the taxonomic ranks and names for the lineage
        lineage_ranks = ncbi.get_rank(lineage)
        lineage_names = ncbi.get_taxid_translator(lineage)
        
        # Extract genus and family
        genus = None
        family = None
        for taxid in lineage:
            if lineage_ranks[taxid] == "genus":
                genus = lineage_names[taxid]
            if lineage_ranks[taxid] == "family":
                family = lineage_names[taxid]
        
        return genus, family
    
    except Exception as e:
        #print(f"An error occurred: {e}")
        return None, None
```

```python
df = pd.read_csv('Species_Level_Host.csv')
```

```python
host = list(df['Host'].values)
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# --- Your Data ---
# This is a placeholder list with more species to demonstrate the grouping.
# Replace it with your actual list of bacteria species.
Host = host


# --- Data Processing and Grouping ---
# 1. Count the occurrences of each species.
species_counts = Counter(Host)

# 2. Identify the top 6 most common species.
# .most_common(6) returns a list of (species, count) tuples.
top_6_species = species_counts.most_common(6)

# 3. Create the labels and sizes for the plot.
labels = [item[0] for item in top_6_species]
sizes = [item[1] for item in top_6_species]

# 4. Group the remaining species into an "Other" category.
# Calculate the total count of the top 6 species.
top_6_count = sum(sizes)
# The "Other" count is the total number of samples minus the top 6 count.
other_count = len(Host) - top_6_count

# 5. Add the "Other" category to our lists if there are any other species.
if other_count > 0:
    labels.append('Other')
    sizes.append(other_count)

# --- Styling (Nature/Science/Cell Inspired) ---
# Use a professional color palette for the main species and a neutral color for "Other".
# Here, we use a colormap for the top species and a distinct grey for "Other".
color_map = plt.cm.viridis_r
main_colors = color_map([i/float(len(labels)-1) for i in range(len(labels)-1)])
colors = list(main_colors)
if other_count > 0:
    colors.append('#cccccc') # A light grey for the "Other" category

# Explode the "Other" slice slightly to indicate it's a collection of species.
explode = [0] * len(top_6_species)
if other_count > 0:
    explode.append(0.1)

# --- Plotting ---
fig, ax = plt.subplots(figsize=(8, 8)) # Create a figure and an axes.

# Create the pie chart
wedges, texts, autotexts = ax.pie(
    sizes,
    explode=explode,
    labels=None, # We will use a legend for labels
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    pctdistance=0.85, # Distance of the percentage text from the center
    wedgeprops=dict(width=0.4, edgecolor='w') # Donut chart style
)

# --- Customization for Publication Quality ---
# Set text properties for clarity
plt.setp(autotexts, size=10, weight="bold", color="white")

# Title and Legend
ax.set_title("Bacterial Species Distribution", fontsize=16, fontweight='bold', pad=20)
ax.legend(
    wedges,
    labels,
    title="Species",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1),
    fontsize=10
)

# Ensure the pie is drawn as a circle.
ax.axis('equal')

# --- Display the Plot ---
plt.tight_layout() # Adjust layout to make room for the legend
plt.show()
fig.savefig("species_pie_chart.pdf", dpi=300, bbox_inches='tight')
# --- Saving the Figure ---
# To save the figure for your paper, uncomment the following line.
# Vector formats (.pdf, .svg) are recommended for publications.
# fig.savefig("top_6_species_pie_chart.pdf", dpi=300, bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np

# --- Your Data ---
# Using the same extended placeholder list.
# Replace it with your actual list of bacteria species.
Host = host

# --- Data Processing ---
# 1. Count the occurrences of each species.
species_counts = Counter(Host)

# 2. Get a list of the abundance values (the counts themselves).
# For our example data, this list will be: [4, 5, 4, 3, 4, 3, 2, 1, 1, 1]
abundance_values = list(species_counts.values())

# --- Plotting ---
fig, ax = plt.subplots(figsize=(8, 6))

# --- Customization for Publication Quality ---
# Define bins for the histogram. Since our counts are integers, we want each bar
# to represent a single integer count. We set bin edges at 0.5, 1.5, 2.5, etc.
# This ensures that all values of '1' fall in one bin, '2' in the next, and so on.
max_abundance = max(abundance_values)
bins = np.arange(0.5, max_abundance + 1.5, 1)

# Create the histogram
ax.hist(abundance_values, bins=bins, color='#34778c', edgecolor='white', linewidth=1.5)

# Set labels and title with a clear, descriptive purpose
ax.set_xlabel('Abundance (Number of Samples per Species)', fontsize=12, fontweight='bold')
ax.set_ylabel('Frequency (Number of Species)', fontsize=12, fontweight='bold')
ax.set_title('Species Abundance Distribution', fontsize=16, fontweight='bold', pad=20)

# Set the x-axis ticks to be integers, corresponding to the counts.
ax.set_xticks(np.arange(1, max_abundance + 1, 1))

# Remove the top and right spines for a cleaner look
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Add a light grid for easier reading, focusing on the y-axis
ax.yaxis.grid(True, linestyle='--', which='major', color='grey', alpha=0.6)
ax.set_axisbelow(True) # Ensure grid is drawn behind the bars

# Adjust layout
plt.tight_layout()

# --- Display the Plot ---
plt.show()

# --- Saving the Figure ---
# To save the figure for your paper, uncomment the following line.
# fig.savefig("species_abundance_histogram.pdf", dpi=300, bbox_inches='tight')
```

```python
cherry = pd.read_csv('cherry_ncbi_results.tsv', sep='\t')
cherry.head()
```

```python
cherry_prediction = {acc: host for acc, host in zip(cherry.Accession, cherry.Host)}
```

```python
cherry_prediction_species = {}
cherry_prediction_genus = {}
```

```python
for acc, host in cherry_prediction.items():
    if 'species:' in host:
        cherry_prediction_species[acc] = host.split('species:')[-1]
    if 'genus:' in host:
        cherry_prediction_genus[acc] = host.split('genus:')[-1]
```

```python
total = 0
correct = 0
for acc in label:
    try:
        pred = cherry_prediction[acc]
        if 'species' not in pred:
            continue
        elif pred.split(':')[1] == label[acc]:
            correct += 1
            total += 1
        else:
            total += 1
    except:
        pass
```
